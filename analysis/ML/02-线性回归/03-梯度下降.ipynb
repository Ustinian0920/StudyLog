```python
# 梯度下降的起始点
xmin = 2
# 梯度下降的上一个点
xmin_last = xmin - 1
# 精度
precision = 0.00001
# 学习率
step = 0.001
# 最大循环次数
max_count = 3000
# 循环次数
count = 0

while True:
    count += 1
    if count >= max_count:
        break
    if np.abs(xmin_last - xmin) <= precision:
        break
    xmin_last = xmin
    xmin = xmin_last - step*g(xmin_last)
    print(f"当前第{count}次迭代，最小值为{xmin}")
```

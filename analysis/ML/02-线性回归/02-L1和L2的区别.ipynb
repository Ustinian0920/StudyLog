```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
sns.set()
colors = sns.color_palette("husl")
```

```python
from sklearn.linear_model import Ridge,Lasso,LinearRegression
```

```python
n_features = 200
n_sample = 50

X = np.random.random(size=(n_sample,n_features))
W = np.random.random(size=200).reshape((-1,1))
```

```python
y = np.dot(X,W)
```

```python
zero_index = np.random.permutation(n_features)[10:]
W[zero_index] = 0
W
```

```python
y_true = np.dot(X,W)
y_true
```

```python
linear = LinearRegression()
# alpha是正则系数
ridge = Ridge(alpha=10) # L2正则 系数会接近0
lasso = Lasso(alpha=0.01) # L1正则 容易产生稀疏矩阵

linear.fit(X,y_true)
ridge.fit(X,y_true)
lasso.fit(X,y_true)
```

```python
fig,axs = plt.subplots(2,2,figsize=(10,6))
fig.tight_layout()
axs[0,0].set_title("True")
axs[0,0].plot(W[:,0],color=colors[0])

axs[0,1].set_title("Linear")
axs[0,1].plot(linear.coef_[0],color=colors[1])

axs[1,0].set_title("Ridge")
axs[1,0].plot(ridge.coef_[0],color=colors[2])

axs[1,1].set_title("Lasso")
axs[1,1].plot(lasso.coef_,color=colors[3])
```

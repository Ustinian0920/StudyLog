```python
# 读取多页
def read_page(doc,start:int,stop:int,step=1) -> list:
    page_list = []
    for page in doc.pages(start,stop,step):
        page_list.append(page)
    return page_list
```

```python
# 读取所有页面
def read_all(doc) -> str:
    str = ""
    for page in doc:
        str += page.get_text("text")
    return str
```

```python
print(read_all(doc))
```
